```python
import numpy as np
from sklearn.base import ClassifierMixin
from sklearn.svm import SVC
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

def train_test() -> None:
    """ Train and test the model """
    digits = load_digits()
    X_train, X_test, y_train, y_test = train_test_split(
        digits.data, digits.target, test_size=0.2
    )

    model = SVC()
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    print(f"Model accuracy: {score}")
    
train_test()
```

```python
from zenml import step
from typing_extensions import Annotated
import pandas as pd
from typing import Tuple

# Importer step
@step
def importer() -> Tuple[
    Annotated[np.ndarray, "X_train"],
    Annotated[np.ndarray, "X_test"],
    Annotated[np.ndarray, "y_train"],
    Annotated[np.ndarray, "y_test"],
]:
    """ Import data """
    digits = load_digits()
    X_train, X_test, y_train, y_test = train_test_split(
        digits.data, digits.target, test_size=0.2
    )
    
    return X_train, X_test, y_train, y_test

# Trainer step
@step
def svc_trainer(X_train: np.ndarray, y_train: np.ndarray) -> ClassifierMixin:
    """ Train SVC model """
    model = SVC()
    model.fit(X_train, y_train)
    
    return model

# Evaluator step
@step
def evaluator(model: ClassifierMixin, X_test: np.ndarray, y_test: np.ndarray) -> float:
    """ Evaluate SVC model """
    score = model.score(X_test, y_test)
    
    return score
```

```python
from zenml import pipeline

# Connect steps to form a pipeline
@pipeline
def digits_pipeline() -> float:
    X_train, X_test, y_train, y_test = importer()
    model = svc_trainer(X_train, y_train)
    evaluator(model, X_test, y_test)
```

```python
digits_svc_pipeline = digits_pipeline()
```
